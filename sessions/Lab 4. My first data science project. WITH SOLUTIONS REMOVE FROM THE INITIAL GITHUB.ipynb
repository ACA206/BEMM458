### Assignment 3. Plot the evolution of life expectancy, population and GDP per capita of Afghanistan

### Assignment 4. Is there any significant relationship between life expectancy and GDP in Afghanistan ?

### Tip: have a look at seaborn-based techniques to represent relationships between two variables.
### check the following link for alternatives: https://seaborn.pydata.org/tutorial.html

```python
ax = sns.kdeplot(data=afghanistan['gdpPercap'],data2=afghanistan['lifeExp'],shade=True) # shade will fill in the contours
ax.set_title('Kernel Density Plot of Total Bill and Tip');
ax.set_xlabel('Total Bill');
```

```python
sns.jointplot("gdpPercap", "lifeExp", data=afghanistan,
                  kind="reg", truncate=False)
```
